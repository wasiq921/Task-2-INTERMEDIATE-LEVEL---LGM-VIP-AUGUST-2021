```python
#import all required libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import plot_tree
from sklearn import tree
```

```python
#import dataset
df = pd.read_csv('/content/sample_data/Iris.csv')
df
```

Successfully Imported and Load our Dataset
and now start Data Analysis

There is 150 rows and 6 Columns in dataset

```python
#splitting Data
fc = [x for x in df.columns if x!="Species"]
x= df[fc]
y= df["Species"]
X_train, X_test, Y_train, Y_test = train_test_split(x,y, random_state = 100, test_size = 0.30)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

Building Decision Tree Model

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
df2 = pd.DataFrame({"Actual":Y_test,"Predicted":Y_pred})
df2.head()
```

Now test Accuracy of Model Prediction

```python
accuracy_score(Y_test,Y_pred)
```

```python
# Plotting
f_n = ["Sepal length", "Sepal width", "Petal length", "Petal width"]
c_n = ["Setosa", "Versicolor", "Virginica"]
plot_tree(model,feature_names = f_n, class_names = c_n , filled = True)
```
